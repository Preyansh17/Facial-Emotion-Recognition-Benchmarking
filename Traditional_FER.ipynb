## Image division

```python
def div(array,m,n):
  arr = np.empty((0,int(array.shape[1]/m),int(array.shape[2]/n)))
  y_train = np.empty(array.shape[0]*32)
  count =0
  for i,img in enumerate(array):
    for r in range(0,int(img.shape[0]-1), int(img.shape[0]/m)):
      for c in range(0,int(img.shape[0]-1 ),int(img.shape[0]/n)):
        arr = np.vstack((arr,img[r:r+int(img.shape[0]/m),c:c+int(img.shape[0]/n)].reshape(1,int(array.shape[1]/m),int(array.shape[2]/n))))
    for i in range(32):
      y_train[i+count] =Y_train[int(count/32)] 
    print(count/32)
    count+=32
  return arr,y_train
```

```python
def stride(array):
  arr = np.empty((0,int(array.shape[1]/5),int(array.shape[2]/5)))
  y_train = np.empty(array.shape[0]*25)
  count=0
  for image in array:
    img_height, img_width = image.shape
    tile_height, tile_width = 19,19

    tiled_array = image.reshape(img_height // tile_height*img_width // tile_width,
                                tile_height,
                                tile_width)
    # tiled_array = tiled_array.swapaxes(1, 2)
    # tiled_array = tiled_array.reshape((tiled_array.shape[0]*tiled_array.shape[1],tiled_array.shape[2],tiled_array.shape[3]))
    arr = np.vstack((arr,tiled_array))
    for i in range(25):
      y_train[i+count] =Y_train[int(count/25)] 
    print(count/25)
    count+=25
    break
  return arr,y_train
```

## Grayscale Extraction

```python
from sklearn.preprocessing import MinMaxScaler
def grayscale(array):
  arr = np.empty((array.shape[0],array.shape[1]*array.shape[2]))
  for i,img in enumerate(array):
    scaler = MinMaxScaler(feature_range=(-1,1))
    img = scaler.fit_transform(img)
    arr[i] = img.ravel()
  return arr
```

```python
from skimage.feature import local_binary_pattern
def lbp_hist(array,p,r):
  arr = np.zeros((array.shape[0],10))
  for i,img in enumerate(array):
    lbp = local_binary_pattern(img, p,r, method="default")
    (hist, _) = np.histogram(lbp.ravel(),
			bins=np.arange(0, p + 3),
			range=(0, p + 2))
    print(hist.shape)
		# normalize the histogram
    hist = hist.astype("float")
    hist /= (hist.sum() + 1e-7)
    arr[i] = hist
		# return the histogram of Local Binary Patterns
  return arr
```

## Gabor Wavelets


```python
import numpy as np
import cv2


 
def build_filters():
 filters = []
 ksize = 31
 sigmaa = [2*(2**(float(m/2))) for m in range(5)]
 for theta in np.arange(0, np.pi, np.pi / 8):
    kern = cv2.getGaborKernel((ksize, ksize), 4, theta, 10, 0.5, 0, ktype=cv2.CV_32F)
    kern /= 1.5*kern.sum()
    filters.append(kern)
 return filters
 
def process(img, filters):
 accum = np.zeros_like(img)
 for kern in filters:
  fimg = cv2.filter2D(img, cv2.CV_8UC3, kern)
  np.maximum(accum, fimg, accum)
 return accum
 
 
filters = build_filters()
arr_1 = np.empty((X_train.shape[0],X_train.shape[1],X_train.shape[2]))
for i,image in enumerate(X_train):
  res1 = process(image, filters)
  arr_1[i] = res1
```
